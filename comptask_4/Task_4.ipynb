```python
import pandas as pd
import math
import numpy as np


lmbda=0.5
l = 1
g = 1
n = 10
u_0 = 0.5
p_0 = 1
T = 2 * math.pi * math.sqrt(l / g)
tau = T / n

u_EE = [u_0]
u_RK = [u_0]
u_AN = [u_0]
           
#Funcs
def f(t, u, p):
    return -p - math.sin(u)

def g(t, u, p):
    return p

#Euler
def Euler(n, tau, u_0, p_0):
    t_n = 0
    u_n = u_0
    p_n = p_0
    for i in range(n):
        u_n += tau * g(t_n, u_n, p_n)
        p_n += tau * f(t_n, u_n, p_n)
        u_EE.append(u_n)
        t_n += tau
    return u_EE

#Runge Kutta
def RungeKutta(n, tau, u_0, p_0):
    t_n = 0
    p_n = p_0
    u_n = u_0

    for i in range(n):
        k_1 = f(t_n, u_n, p_n)
        q_1 = g(t_n, u_n, p_n)

        k_2 = f(t_n + tau / 2, u_n + (tau / 2) * q_1, p_n + (tau / 2) * k_1)
        q_2 = g(t_n + tau / 2, u_n + (tau / 2) * q_1, p_n + (tau / 2) * k_1)

        k_3 = f(t_n + tau / 2, u_n + (tau / 2) * q_2, p_n + (tau / 2) * k_2)
        q_3 = g(t_n + tau / 2, u_n + (tau / 2) * q_2, p_n + (tau / 2) * k_2)

        k_4 = f(t_n + tau, u_n + tau * q_3, p_n + tau * k_3)
        q_4 = g(t_n + tau, u_n + tau * q_3, p_n + tau * k_3)

        p_n = p_n + tau * (k_1 + 2 * k_2 + 2 * k_3 + k_4)/6
        u_n = u_n + tau * (q_1 + 2 * q_2 + 2 * q_3 + q_4)/6

        u_RK.append(u_n)
        
        t_n += tau
    return u_RK

u_EE = Euler(n, tau, u_0, p_0)
u_RK = RungeKutta(n, tau, u_0, p_0) 

index = np.linspace(0, T, n+1)
data = {"u_EE" : u_EE, "u_RK" : u_RK}
table = pd.DataFrame(data, index=index)

print(table)
```
